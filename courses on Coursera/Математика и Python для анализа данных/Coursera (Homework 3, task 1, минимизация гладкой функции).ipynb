## Минимизация гладкой функции

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
x = np.arange(1.,30.1, 0.1)
def f(x):
    return np.sin(x / 5.) * np.exp(x / 10.) + 5. * np.exp(-x / 2.)
```

```python
plt.figure(figsize = (9,6))
plt.plot(x, f(x), color = 'green', label = 'Исходная функция')
plt.xlabel('x', fontsize = 12, fontweight = 'bold')
plt.ylabel('f(x)', fontsize = 12, rotation = 0, fontweight = 'bold')
plt.grid(True, alpha = 0.6, color = 'black')
plt.legend()
plt.show()
```

### Начиная минимизировать функцию в промежутке x от 1 до 10 включительно, мы выходим на локальный минимум в точке x = 4.1363, далее в промежутке с x = 11 и по 29 нас выкидывает в другой минимум в точке x = 25.88, но в точке x = 13 нас выкидывает за промежутки нашего рассмотрения и попадаем в минимум в точке x = 88.71. 

```python
for i in range(1,31,1):
    result_2 = minimize(f, [i])
    print ('\n Значение x =',i,'\n','\n', result_2,'\n\n')
```

```python
points = [2.,30.]
Results = []
for i in points:   
    results = minimize(f,[i], method = 'BFGS')
    Results.append(round(results.fun,2))
    print ('\n Значение x =',i,'\n','\n', results,'\n\n')
print (Results)
```

### Точка x = 4.14 - это координата локального минимума, а точка x = 25.88 - координата глобального минимума на отрезке [1,30]. Запишем полученные результаты в текстовый файл.

```python
with open('D:\Coursera\Homework3(Task1).txt', 'w') as file:
    file.writelines(Results+ ' ' for Results in map(str, Results))
file.close()
```
